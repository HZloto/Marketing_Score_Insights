```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/hgoz/Desktop/ESADE/MASTER/CAPSTONE/MARCO DATASETS/CODE/df_merged.csv")
```

```python
df = df.drop(['Unnamed: 0', 'conversions_ads',
        'engagements_ads', 
       'interactions_ads', 'interaction_rate_ads', 'average_cpm_ads',
       'ctr_ads', 'average_cpc_ads', 'conversions_value_ads',
       'value_per_conversion_ads', 'engagement_rate_ads', 'average_cost_ads',
       'cost_per_conversion_ads', 'average_cpe_ads',
        'impressions_anal',
       'adClicks_anal'], axis = 1)
```

```python
df['date'] = pd.to_datetime(df['date'])
weekly_summary = df.resample('W', on='date').sum()
monthly_summary = df.resample('M', on='date').sum()
```

```python
df_merged_week = weekly_summary
df_merged_week.to_csv("df_merged_week_interest.csv")

df_merged_month = monthly_summary
df_merged_month.to_csv("df_merged_month_interest.csv")

```
